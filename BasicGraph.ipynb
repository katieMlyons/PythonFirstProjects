```python
# making a basic line graph

from bokeh.plotting import figure, output_file, show

x = [1,2,3,4,5]
y = [6,7,8,9,10]

output_file('line.html')

f=figure()

f.line(x,y)
show(f)
```

```python
# making a basic line graph

from bokeh.plotting import figure, output_file, show
import pandas as pd

df=pd.read_csv('data.csv')
x=df['x']
y=df['y']

output_file('line.html')

f=figure()

f.line(x,y)
show(f)
```

```python
p=figure(plot_width=500,plot_height=400, tools='pan') #logo=None

p.line([1,2,3],[4,5,6])
p.title.text="Cool Data"
p.title.text_color="gray"
p.title.text_font="times"
p.title.text_font_style="bold"
p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color=None
p.xaxis.axis_label="Date"
p.yaxis.axis_label="Intensity"    
 
output_file("graph.html")
show(p)
```

```python
weather = pd.read_excel('http://pythonhow.com/verlegenhuken.xlsx')
weather.columns
```

```python
temp = weather.Temperature / 10
pres = weather.Pressure / 10
```

```python
w = figure(plot_width=500,plot_height=400,tools='pan')

w.title.text = "Temperature and Air Pressure"
w.title.text_color = "gray"
w.title.text_font="arial"
w.title.text_font_style="bold"
w.xaxis.minor_tick_line_color=None
w.yaxis.minor_tick_line_color=None
w.xaxis.axis_label='Temperature (' + '\u2103' + ')'
w.yaxis.axis_label="Pressure (hPa)"    
w.circle(temp,pres, size=0.5)

output_file('weather.html')
show(w)
```

```python
from bokeh.plotting import figure, output_file, show
p = figure(plot_width=400, plot_height=500)
p.circle([1,2,3,4,5], [5,6,5,5,3], size = [i*2 for i in [8,12,14,15,20]], color="red", alpha=0.5)

p.title.text = "Earthquakes"
p.title.text_color = "orange"
p.title.text_font = "times"
p.title.text_font_style = "italic"
p.yaxis.minor_tick_line_color = "Yellow"
p.xaxis.axis_label = "Times"
p.yaxis.axis_label = "Value"

output_file("Scatter_plotting.html")
show(p)
```

```python
from bokeh.plotting import figure, output_file, show

p = figure(plot_width=300, plot_height=300)
p.circle([1,2], [3,4])

# configure visual properties on a plot's title attribute
p.title.text = "Title With Options"
p.title.align = "right"
p.title.text_color = "orange"
p.title.text_font_size = "25px"
p.title.background_fill_color = "#aaaaee"

output_file("title.html")

show(p)
```
